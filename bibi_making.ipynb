```python
# 网格交易法则
security = ['eth_usd','btc_usd']
#period = ['this_week','next_week','quarter']
symbol ='eth_usd'
contractType='this_week'
unit_volume = 5
timeframe = '1hour'
#三天
limit = '72'
TREND = True
STOCK = False
UP    = True
DOWN  = False
context.trend_direction = UP
context.price = context.future_ticker(symbol,contractType)['ticker']['last']
context.last_price = context.price
context.volume= 20
context.status = STOCK
def price_to_df(symbol,contractType,frequency,count):
    #print(symbol,contractType,frequency,count)
    price = context.future_kline(symbol,contractType,type=frequency,size=count)
    #print(price)
    df = pd.DataFrame(columns  = ['timestamp','open','high','low','close','volume','coin_amount'])
    i=0
    for k in price:
        df.loc[i,:]=k
        i = i+1
    return df
# end price is better than start_price
def adjust_network(context,high_price,low_price):
    if(low_price>context.price):
        for i in range(int(low_price),int(high_price),5):
            context.close_long(symbol,contractType,i,context.volume,0,20)
    elif(high_price<context.price):
        for i in range(int(low_price),int(high_price),5):
            context.close_short(symbol,contractType,i,context.volume,0,20)
    else:
        for i in range(int(low_price),int(context.price),5):
            context.close_short(symbol,contractType,i,context.volume,0,20)
        for i in range(int(context.price)+5,int(high_price),5):
            context.close_long(symbol,contractType,i,context.volume,0,20)
        #cof = 20*(context.high-context.price)/(context.high-context.low)
        #volume  =int((10-abs(cof-10))*unit_volume)
    context.network_up = max(high_price,context.network_up)
    context.network_down=min(low_price,context.network_down)
def net_work_trading(context):
    count=0
    #context.network_up = 1250
    #context.network_down=1100
    while True:
        count = count+1
        pf = context.get_all_future_position(['btc_usd','eth_usd'],['this_week','next_week','quarter'])
        context.long_volume = pf.loc[3,'buy_amount']
        context.short_volume = pf.loc[3,'sell_amount']
        #print(context.long_volume,context.short_volume)
        
        if context.price>(context.network_up*0.998):
            print(' too upp, add network%f %f' %(context.network_up,context.network_down))
            adjust_network(context,context.network_up*1.05,context.network_up)
            #context.network_up = context.network_up*1.05
        if context.price<(context.network_down*1.002):
            print('too low add network%f %f' %(context.network_up,context.network_down))
            adjust_network(context,context.network_down,context.network_down*0.95)
            #contxt.network_down = context.network_down*0.95
     #       add_network(context,start,end_price)
        #okcoinFuture.get_all_future_position(security,period)
        context.tick = context.future_ticker(symbol,contractType)['ticker']
        context.price= context.tick['last']
        context.bar = price_to_df(symbol,contractType,frequency=timeframe,count=72)
        #print(context.bar)
        context.high = max(context.bar['high'])
        context.low  = min(context.bar['low'])
        #print(type(context.price))
        #print(context.high)
        # 价格越高， cof 越小。
        context.cof = 20*(context.high-context.price)/(context.high-context.low)
        cof = context.cof
        context.volume  =int((10-abs(cof-10))*unit_volume)
        volume = context.volume
        #价格越靠近边界，仓位越轻
        #价格在上方，
        #if cof>0.75:
        # 越接近两端，volume 越小
        #print('high:',context.high,'low:',context.low,'price:',context.price,volume)
        #print(price)
        if (context.price>context.last_price*1.004):
            # 向上突破一个单元格
            context.open_short(symbol,contractType,price=context.price,amount=volume,matchType=0,leverRate=20)
            context.close_short(symbol,contractType,price=context.last_price,amount=volume,matchType=0,leverRate=20)
            #context.open_long(symbol,contractType,price=context.last_price,amount=volume*2,)
            # 仓压太大，调节一下。
            if(context.long_volume<(context.short_volume*0.5)):
                # 开几个多仓。
                context.open_long(symbol,contractType,price=context.price,amount=volume*2,matchType=0,leverRate=20)
            #    context.open_long(symbol,contractType,price=context.tick['last'],amount=volume*5,matchType=0,leverRate=20)
            context.last_price = context.price
        if (context.price<context.last_price*0.996):
            # 向上突破一个单元格
            context.open_long(symbol,contractType,price=context.price,amount=volume,matchType=0,leverRate=20)
            context.close_long(symbol,contractType,price=context.last_price,amount=volume,matchType=0,leverRate=20)    
            # 仓压太大，调节一下。
            if(context.long_volume>(context.short_volume*2)):
                context.open_short(symbol,contractType,price=context.tick['last'],amount=volume*2,matchType=0,leverRate=20)
            context.last_price = context.price
        #  进入趋势期: 返回趋势期:
        if(context.status == STOCK and context.price>1305):
            # 价格突破 三日最高点: 向上趋势
            context.status = TREND
            context.trend_direction = UP
            context.close_short(symbol,contractType,price=context.tick['last'],amount=1000,matchType=0,leverRate=20)
            context.open_long(symbol,contractType,price=context.tick['last'],amount=2000,matchType=0,leverRate=20
        if(context.status == TREND and context.trend_direction == UP and context.price<1305):
            context.open_short(symbol,contractType,price=context.tick['last'],amount=1000,matchType=0,leverRate=20)
            context.close_long(symbol,contractType,price=context.tick['last'],amount=2000,matchType=0,leverRate=20)
            context.status = STOCK
        sleep(5)
    return 0
context.network_up = 1425
context.network_down = 1100
#adjust_network(context,context.network_up,context.network_down)
while True:
    try:
        net_work_trading(context)
    except:
        pass
```
